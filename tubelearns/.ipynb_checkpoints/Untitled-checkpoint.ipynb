```python

with open("./raw_data/raw_data.txt" , "r") as f:
    content = f.read()
```

```python
from tokenizers import Cleaning , Tokenization

Cleaner = Cleaning()
Tokenizer = Tokenization()

Sentence_Tokens = Tokenizer.tokenize_rawlower(content)
```

```python
Cleaning_Tokens = Cleaner.punct_list(Sentence_Tokens)
```

```python
Cleaning_2 = Cleaner.lemmatizer(Cleaning_Tokens)
```

```python
words = ["and","or","but"]

def conjunct_split_list(sentences, conjunctions ):
    splitted = []

    for test in sentences:
        for conjunction in conjunctions:
            print(conjunction,"|" , test,"\n")
            while conjunction == test:
                test_final = test.split(conjunction, 1)
                part1 = test_final[0].strip()
                part2 = test_final[1].strip()
                splitted.append(part1)
                test = part2            
        if test.strip():
            splitted.append(test.strip())

    return splitted


Cleaning_3 = conjunct_split_list(Cleaning_2,words)

```
